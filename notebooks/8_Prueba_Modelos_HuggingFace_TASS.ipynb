## 4. PARTE 1: Ejemplos TASS (Sentimientos)

Ejemplos variados para probar los 3 sentimientos (Positivo, Neutral, Negativo):

````python
# Ejemplos de prueba con etiqueta esperada
test_examples = [
    # Sentimientos POSITIVOS
    {
        'texto': '¡Me encanta este producto! Superó todas mis expectativas 😊',
        'esperado': 'P'
    },
    {
        'texto': 'Qué día tan hermoso, el mejor de la semana',
        'esperado': 'P'
    },
    {
        'texto': 'Excelente servicio, muy recomendado. Volveré sin duda',
        'esperado': 'P'
    },
    
    # Sentimientos NEGATIVOS
    {
        'texto': 'Pésimo servicio al cliente. Nunca me respondieron',
        'esperado': 'N'
    },
    {
        'texto': 'Qué decepción, esperaba mucho más de este lugar',
        'esperado': 'N'
    },
    {
        'texto': 'No lo recomiendo para nada. Mala experiencia 😠',
        'esperado': 'N'
    },
    
    # Sentimientos NEUTRALES
    {
        'texto': 'El producto llegó en 3 días, como indicaba el vendedor',
        'esperado': 'NEU'
    },
    {
        'texto': 'Es un lugar normal, ni muy bueno ni muy malo',
        'esperado': 'NEU'
    },
    {
        'texto': 'La reunión fue a las 10 am según lo planeado',
        'esperado': 'NEU'
    },
    
    # Casos AMBIGUOS (difíciles)
    {
        'texto': 'El precio está bien, aunque la calidad podría mejorar',
        'esperado': 'NEU'  # Mixto: aspecto positivo y negativo
    },
    {
        'texto': 'Esperaba más pero tampoco está tan mal',
        'esperado': 'NEU'  # Opinión balanceada
    }
]

# Mapeo de etiquetas
label_names = {
    'P': 'POSITIVO',
    'N': 'NEGATIVO',
    'NEU': 'NEUTRAL'
}

print(f'📝 {len(test_examples)} ejemplos de prueba cargados')
print(f'   - Positivos: {sum(1 for e in test_examples if e["esperado"] == "P")}')
print(f'   - Negativos: {sum(1 for e in test_examples if e["esperado"] == "N")}')
print(f'   - Neutrales: {sum(1 for e in test_examples if e["esperado"] == "NEU")}')
````

## 5. Funciones de Predicción (TASS - 3 clases)

````python
def predict_sentiment(pipe, text, model_name='Modelo'):
    """
    Predice el sentimiento de un texto (Positivo, Neutral, Negativo)
    
    Args:
        pipe: Pipeline de HuggingFace
        text: Texto a clasificar
        model_name: Nombre del modelo para mostrar
    
    Returns:
        dict con predicción y probabilidades
    """
    # Obtener predicción (pipeline retorna lista de resultados)
    output = pipe(text)
    
    # Manejar formato de salida (puede ser [[{...}]] o [{...}])
    if isinstance(output[0], list):
        results = output[0]  # return_all_scores=True retorna [[{...}]]
    else:
        results = output  # Formato alternativo [{...}]
    
    # Extraer probabilidades para las 3 clases
    probs = {}
    for r in results:
        label = r['label']
        score = r['score'] * 100
        
        # Mapear LABEL_X a etiquetas legibles
        if label == 'LABEL_0':
            probs['N'] = score
        elif label == 'LABEL_1':
            probs['NEU'] = score
        elif label == 'LABEL_2':
            probs['P'] = score
        else:
            probs[label] = score
    
    # Determinar predicción (clase con mayor probabilidad)
    prediccion = max(probs, key=probs.get)
    confianza = probs[prediccion]
    
    return {
        'modelo': model_name,
        'prediccion': prediccion,
        'prediccion_nombre': label_names[prediccion],
        'confianza': confianza,
        'prob_positivo': probs.get('P', 0),
        'prob_neutral': probs.get('NEU', 0),
        'prob_negativo': probs.get('N', 0)
    }


def print_prediction(text, result, esperado=None):
    """
    Imprime resultados de predicción con formato bonito
    """
    # Emoji según predicción
    emojis = {'P': '😊', 'N': '😠', 'NEU': '😐'}
    emoji = emojis.get(result['prediccion'], '❓')
    
    # Color de resultado (simulado con emojis)
    if esperado:
        match = '✅' if result['prediccion'] == esperado else '❌'
    else:
        match = ''
    
    print(f"\n{'='*80}")
    print(f"📝 Texto: {text}")
    print(f"{'='*80}")
    print(f"🤖 Modelo: {result['modelo']}")
    print(f"🎯 Predicción: {result['prediccion_nombre']} ({result['prediccion']}) {emoji} {match}")
    print(f"📊 Confianza: {result['confianza']:.2f}%")
    print(f"   • Prob. Positivo: {result['prob_positivo']:.2f}%")
    print(f"   • Prob. Neutral: {result['prob_neutral']:.2f}%")
    print(f"   • Prob. Negativo: {result['prob_negativo']:.2f}%")
    if esperado:
        print(f"✓ Esperado: {label_names[esperado]} ({esperado})")

print('✅ Funciones de predicción definidas')
````

## 6. Probar Modelos TASS - BETO (F1: 64.54%) ⭐

````python
print('\n' + '='*80)
print('🔵 PROBANDO MODELO: BETO-TASS (dccuchile/bert-base-spanish-wwm-cased)')
print(f'   📍 HuggingFace: {MODEL_BETO_TASS}')
print(f'   🎯 F1-Score Test: 64.54%')
print('='*80)

beto_tass_results = []

for i, example in enumerate(test_examples, 1):
    print(f"\n--- Ejemplo {i}/{len(test_examples)} ---")
    
    result = predict_sentiment(
        pipe_beto_tass,
        example['texto'],
        model_name='BETO-TASS'
    )
    
    print_prediction(
        example['texto'],
        result,
        esperado=example['esperado']
    )
    
    beto_tass_results.append(result)

# Calcular accuracy
correct_beto_tass = sum(1 for i, r in enumerate(beto_tass_results) 
                   if r['prediccion'] == test_examples[i]['esperado'])
accuracy_beto_tass = (correct_beto_tass / len(test_examples)) * 100

print(f"\n\n{'='*80}")
print(f"📊 RESUMEN BETO-TASS:")
print(f"   ✅ Correctas: {correct_beto_tass}/{len(test_examples)}")
print(f"   🎯 Accuracy: {accuracy_beto_tass:.2f}%")
print(f"   📈 F1-Score (test original): 64.54%")
print(f"{'='*80}")
````

## 7. Probar Modelos TASS - XLM-RoBERTa (F1: 64.41%)

````python
print('\n' + '='*80)
print('🟣 PROBANDO MODELO: XLM-RoBERTa-TASS (xlm-roberta-base)')
print(f'   📍 HuggingFace: {MODEL_XLMR_TASS}')
print(f'   🎯 F1-Score Test: 64.41%')
print('='*80)

xlmr_tass_results = []

for i, example in enumerate(test_examples, 1):
    print(f"\n--- Ejemplo {i}/{len(test_examples)} ---")
    
    result = predict_sentiment(
        pipe_xlmr_tass,
        example['texto'],
        model_name='XLM-RoBERTa-TASS'
    )
    
    print_prediction(
        example['texto'],
        result,
        esperado=example['esperado']
    )
    
    xlmr_tass_results.append(result)

# Calcular accuracy
correct_xlmr_tass = sum(1 for i, r in enumerate(xlmr_tass_results) 
                   if r['prediccion'] == test_examples[i]['esperado'])
accuracy_xlmr_tass = (correct_xlmr_tass / len(test_examples)) * 100

print(f"\n\n{'='*80}")
print(f"📊 RESUMEN XLM-RoBERTa-TASS:")
print(f"   ✅ Correctas: {correct_xlmr_tass}/{len(test_examples)}")
print(f"   🎯 Accuracy: {accuracy_xlmr_tass:.2f}%")
print(f"   📈 F1-Score (test original): 64.41%")
print(f"{'='*80}")
````

````python
# Ejemplos de sarcasmo con etiqueta esperada
sarcasmo_examples = [
    {
        'texto': '¡Qué divertido, otro lunes de trabajo! 🙃',
        'esperado': 'SARCASMO'
    },
    {
        'texto': 'Me encanta cuando llueve el día que tengo planes al aire libre',
        'esperado': 'SARCASMO'
    },
    {
        'texto': 'Obvio que aprobé el examen, estudié exactamente 5 minutos',
        'esperado': 'SARCASMO'
    },
    {
        'texto': 'Qué sorpresa que el internet se cayó justo cuando iba a entregar la tarea',
        'esperado': 'SARCASMO'
    },
    {
        'texto': 'Sí, porque claramente yo soy el experto en cocina gourmet 🍳',
        'esperado': 'SARCASMO'
    },
    {
        'texto': 'Hoy tuve un día maravilloso en el parque con mi familia',
        'esperado': 'NO_SARCASMO'
    },
    {
        'texto': 'Estoy muy contento con los resultados del proyecto',
        'esperado': 'NO_SARCASMO'
    },
    {
        'texto': 'El curso me pareció muy útil e interesante',
        'esperado': 'NO_SARCASMO'
    }
]

print(f'📝 {len(sarcasmo_examples)} ejemplos de sarcasmo cargados')
print(f'   - Sarcasmo: {sum(1 for e in sarcasmo_examples if e["esperado"] == "SARCASMO")}')
print(f'   - No Sarcasmo: {sum(1 for e in sarcasmo_examples if e["esperado"] == "NO_SARCASMO")}')
````

## 10. Funciones de Predicción (Sarcasmo - 2 clases)

````python
def predict_sarcasm(pipe, text, model_name='Modelo'):
    """
    Predice si un texto es sarcástico o no (clasificación binaria)
    """
    # Obtener predicción (pipeline retorna lista de resultados)
    output = pipe(text)
    
    # Manejar formato de salida (puede ser [[{...}]] o [{...}])
    if isinstance(output[0], list):
        results = output[0]  # return_all_scores=True retorna [[{...}]]
    else:
        results = output  # Formato alternativo [{...}]
    
    # Extraer probabilidades
    probs = {}
    for r in results:
        label = r['label']
        score = r['score'] * 100
        
        # Mapear LABEL_X a etiquetas legibles
        if label == 'LABEL_0':
            probs['NO_SARCASMO'] = score
        elif label == 'LABEL_1':
            probs['SARCASMO'] = score
        else:
            probs[label] = score
    
    prediccion = max(probs, key=probs.get)
    confianza = probs[prediccion]
    
    return {
        'modelo': model_name,
        'prediccion': prediccion,
        'confianza': confianza,
        'prob_sarcasmo': probs.get('SARCASMO', 0),
        'prob_no_sarcasmo': probs.get('NO_SARCASMO', 0)
    }


def print_sarcasm_prediction(text, result, esperado=None):
    """
    Imprime resultados de predicción de sarcasmo
    """
    emoji = '😏' if result['prediccion'] == 'SARCASMO' else '😊'
    
    if esperado:
        match = '✅' if result['prediccion'] == esperado else '❌'
    else:
        match = ''
    
    print(f"\n{'='*80}")
    print(f"📝 Texto: {text}")
    print(f"{'='*80}")
    print(f"🤖 Modelo: {result['modelo']}")
    print(f"🎯 Predicción: {result['prediccion']} {emoji} {match}")
    print(f"📊 Confianza: {result['confianza']:.2f}%")
    print(f"   • Prob. Sarcasmo: {result['prob_sarcasmo']:.2f}%")
    print(f"   • Prob. No Sarcasmo: {result['prob_no_sarcasmo']:.2f}%")
    if esperado:
        print(f"✓ Esperado: {esperado}")

print('✅ Funciones de predicción de sarcasmo definidas')
````

## 11. Probar Modelos Sarcasmo - BETO (F1: 99.56%) ⭐⭐

````python
print('\n' + '='*80)
print('🟢 PROBANDO MODELO: BETO-Sarcasmo (dccuchile/bert-base-spanish-wwm-cased)')
print(f'   📍 HuggingFace: {MODEL_BETO_SARCASMO}')
print(f'   🎯 F1-Score Test: 99.56%')
print('='*80)

beto_sarcasmo_results = []

for i, example in enumerate(sarcasmo_examples, 1):
    print(f"\n--- Ejemplo {i}/{len(sarcasmo_examples)} ---")
    
    result = predict_sarcasm(
        pipe_beto_sarcasmo,
        example['texto'],
        model_name='BETO-Sarcasmo'
    )
    
    print_sarcasm_prediction(
        example['texto'],
        result,
        esperado=example['esperado']
    )
    
    beto_sarcasmo_results.append(result)

# Calcular accuracy
correct_beto_sarcasmo = sum(1 for i, r in enumerate(beto_sarcasmo_results) 
                             if r['prediccion'] == sarcasmo_examples[i]['esperado'])
accuracy_beto_sarcasmo = (correct_beto_sarcasmo / len(sarcasmo_examples)) * 100

print(f"\n\n{'='*80}")
print(f"📊 RESUMEN BETO-Sarcasmo:")
print(f"   ✅ Correctas: {correct_beto_sarcasmo}/{len(sarcasmo_examples)}")
print(f"   🎯 Accuracy: {accuracy_beto_sarcasmo:.2f}%")
print(f"   📈 F1-Score (test original): 99.56%")
print(f"{'='*80}")
````
