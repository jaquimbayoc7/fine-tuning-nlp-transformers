## 3. Prompts Templates

```python
PROMPT_ZERO_SHOT = """Eres un asistente mÃ©dico experto en oncologÃ­a especializado en cÃ¡ncer de prÃ³stata.

Tu tarea es extraer informaciÃ³n clÃ­nica estructurada del siguiente texto mÃ©dico.

ENTIDADES A EXTRAER:
1. edad: Edad del paciente
2. sexo: Sexo del paciente
3. biomarcadores: PSA, Gleason, TNM
4. tipo_cancer: Tipo especÃ­fico de cÃ¡ncer
5. cirugias: Procedimientos quirÃºrgicos
6. diagnostico: DiagnÃ³stico principal
7. imagenes: Hallazgos de estudios por imagen
8. examenes: ExÃ¡menes realizados
9. antecedentes: Antecedentes mÃ©dicos
10. tratamiento: Tratamientos aplicados
11. medicacion: Medicamentos prescritos

Responde ÃšNICAMENTE con un objeto JSON vÃ¡lido. NO agregues explicaciones.

Texto: {texto}

JSON:"""

PROMPT_FEW_SHOT = """Eres un asistente mÃ©dico experto. Extrae entidades clÃ­nicas en formato JSON.

Ejemplo:
Texto: "VarÃ³n de 58 aÃ±os. PSA 12.5 ng/dL. Gleason 4+3. Adenocarcinoma prostÃ¡tico."
JSON: {{"edad": ["58 aÃ±os"], "sexo": ["masculino"], "biomarcadores": ["PSA 12.5 ng/dL", "Gleason 4+3"], "tipo_cancer": ["adenocarcinoma prostÃ¡tico"]}}

Ahora extrae del siguiente texto:
Texto: {texto}

JSON:"""

PROMPTS = {'v1': PROMPT_ZERO_SHOT, 'v2': PROMPT_FEW_SHOT}
print('âœ“ Prompts cargados: v1 (zero-shot), v2 (few-shot)')
```

## 4. FunciÃ³n de Inferencia via API y MÃ©tricas

```python
import json, re, time
from tqdm import tqdm

def call_groq_api(groq_model, user_content, max_tokens=512, retries=3):
    for attempt in range(retries):
        try:
            response = client_groq.chat.completions.create(
                model=groq_model,
                messages=[
                    {"role": "system", "content": "Eres un asistente mÃ©dico experto en oncologÃ­a. Extrae entidades clÃ­nicas del texto y responde SIEMPRE en JSON vÃ¡lido."},
                    {"role": "user", "content": user_content}
                ],
                max_tokens=max_tokens,
                temperature=0.1
            )
            return response.choices[0].message.content
        except Exception as e:
            print(f"  âš ï¸ Intento {attempt+1} fallido: {e}")
            time.sleep(2 ** attempt)
    return None

def parse_json_from_response(response):
    if not response:
        return {}
    # Buscar bloque JSON en la respuesta
    match = re.search(r'\{.*\}', response, re.DOTALL)
    if match:
        try:
            return json.loads(match.group())
        except:
            pass
    # Intentar parsear directamente
    try:
        return json.loads(response)
    except:
        pass
    return {}  # â† Devuelve dict vacÃ­o en vez de None

def normalizar_a_lista(v):
    """Convierte cualquier valor a lista iterable."""
    if v is None:
        return []
    if isinstance(v, list):
        return v
    return [v]  # int, str, float â†’ envuelve en lista

def calcular_metricas_globales(extraidas, esperadas):
    if not extraidas or not isinstance(extraidas, dict):
        extraidas = {}
    todas = set(list(esperadas.keys()) + list(extraidas.keys()))
    tp_total, fp_total, fn_total = 0, 0, 0
    for cat in todas:
        esp = set(str(e).lower() for e in normalizar_a_lista(esperadas.get(cat, [])))
        ext = set(str(e).lower() for e in normalizar_a_lista(extraidas.get(cat, [])))
        if not esp and not ext:
            continue
        tp_total += len(esp & ext)
        fp_total += len(ext - esp)
        fn_total += len(esp - ext)
    precision = tp_total / (tp_total + fp_total) if (tp_total + fp_total) > 0 else 0
    recall    = tp_total / (tp_total + fn_total) if (tp_total + fn_total) > 0 else 0
    f1        = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0
    return {'precision': precision, 'recall': recall, 'f1': f1,
            'tp': tp_total, 'fp': fp_total, 'fn': fn_total}

def evaluar_modelo(nombre_corto, groq_model):
    print('='*60)
    print(f'ðŸš€ Evaluando: {nombre_corto}')
    print(f'   Groq model: {groq_model}')
    print('='*60)
    resultados = []
    for version, template in PROMPTS.items():
        print(f'\n  Prompt {version}:')
        for texto_data in tqdm(textos_eval, desc=f'{nombre_corto}-{version}'):
            user_content = template.format(texto=texto_data['texto'])
            respuesta = call_groq_api(groq_model, user_content)
            entidades = parse_json_from_response(respuesta)
            metricas  = calcular_metricas_globales(entidades, texto_data['entidades'])
            resultados.append({
                'modelo': nombre_corto,
                'version_prompt': version,
                'texto_id': texto_data.get('texto_id', 'unknown'),
                'metricas': metricas,
                'entidades_extraidas': entidades,
                'entidades_esperadas': texto_data['entidades'],
                'respuesta_raw': respuesta
            })
            time.sleep(0.5)  # evitar rate limit
    return resultados
```

## 5. Evaluar los 4 Modelos

> **Tiempo estimado**: ~10-15 minutos (Groq es extremadamente rÃ¡pido)
> 
> Si algÃºn modelo falla por **503 (model loading)**, espera 20 segundos y reejecutar esa celda â€” los modelos se cargan on-demand en los servidores de HuggingFace.

```python
# Modelo 1: Llama-3.3-70B (grande, alta calidad)
resultados_m1 = evaluar_modelo('Llama-3.3-70B', MODELOS_GROQ['Llama-3.3-70B'])
```

```python
# Modelo 2: Qwen3-32B (familia Qwen, arquitectura diferente)
resultados_m2 = evaluar_modelo('Qwen3-32B', MODELOS_GROQ['Qwen3-32B'])
```

```python
# Modelo 3: Llama-4-Scout-17B (mediano, eficiente)
resultados_m3 = evaluar_modelo('Llama4-Scout-17B', MODELOS_GROQ['Llama4-Scout-17B'])
```

```python
# Modelo 4: Llama-3.1-8B (pequeÃ±o, rÃ¡pido)
resultados_m4 = evaluar_modelo('Llama-3.1-8B', MODELOS_GROQ['Llama-3.1-8B'])
```

## 6. AnÃ¡lisis de Resultados

```python
todos_resultados = resultados_m1 + resultados_m2 + resultados_m3 + resultados_m4

df_rows = []
for r in todos_resultados:
    if 'metricas' in r and 'f1' in r['metricas']:
        df_rows.append({
            'Modelo': r['modelo'],
            'Prompt': r['version_prompt'],
            'Texto_ID': r['texto_id'],
            'Precision': r['metricas']['precision'],
            'Recall': r['metricas']['recall'],
            'F1': r['metricas']['f1']
        })

df = pd.DataFrame(df_rows)

resumen = df.groupby(['Modelo', 'Prompt']).agg(
    F1_Mean=('F1', 'mean'),
    F1_Std=('F1', 'std'),
    Precision=('Precision', 'mean'),
    Recall=('Recall', 'mean'),
    Textos=('Texto_ID', 'count')
).round(4).reset_index()

print('='*80)
print('ðŸ“Š RESULTADOS DE PROMPT ENGINEERING (Groq API)')
print('='*80)
print(resumen.to_string(index=False))
print('='*80)
print('ðŸŽ¯ COMPARACIÃ“N CON FINE-TUNING:')
print('  BETO  (fine-tuned): 96.18% F1')
print('  XLM-R (fine-tuned): 96.70% F1')
print('='*80)
```

## 7. Guardar Resultados

```python
resumen.to_csv('resultados_prompt_engineering.csv', index=False)

with open('resultados_detallados.json', 'w', encoding='utf-8') as f:
    json.dump({
        'Llama-3.3-70B': resultados_m1,
        'Qwen3-32B': resultados_m2,
        'Llama-4-Scout-17B': resultados_m3,
        'Llama-3.1-8B': resultados_m4
    }, f, ensure_ascii=False, indent=2)

print('âœ… Archivos guardados:')
print('  - resultados_prompt_engineering.csv')
print('  - resultados_detallados.json')
```
